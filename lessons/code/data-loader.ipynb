### Custom Reader

```python
from llama_index.core.readers.base import BaseReader
from llama_index.core import Document
```

```python
class MyQAReader(BaseReader):
    """
    It aims to parse the content in qa file and build documents for llamaIndex
    """
    def load_data(self, file, extra_info=None):
        with open(file, "r") as f:
            text = f.read()
        qa_list = map(lambda x:x.strip(), text.split("---"))  # Each Q&A seperate with "---"
        return [Document(text=qa, extra_info=extra_info or {}) for qa in qa_list if qa != ""]
```

```python
from llama_index.core import SimpleDirectoryReader

reader = MyQAReader()
file_extractor = {".qa": reader}
documents = SimpleDirectoryReader("../data/qa/", file_extractor=file_extractor).load_data()
```
